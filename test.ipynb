```python
from db import get_connection

try:
    conn = get_connection()
    cursor = conn.cursor()
    cursor.execute("SELECT version();")
    version = cursor.fetchone()
    print("Соединение с БД успешно:", version)
    cursor.close()
    conn.close()
except Exception as e:
    print("Ошибка подключения к БД:", e)
```
